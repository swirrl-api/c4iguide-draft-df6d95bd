# Welcome to Climat4Impact v2 *(alpha)*


This new version of the C4I portal is a complete redesing of the old website (still available at http://climate4impact.eu). 

In v2 we improved the search and data analysis functionalities, aiming at supporting climate change impact researchers with more flexibility in the way they access and manipulate the data provided by the Earth System Grid Federation (ESGF). 

In this page we provide a short guide to the use of C4I v2. 


## Do I need an account to use C4I v2?


You can access the portal as **guest** or as **ESGF Registred User**. On the home page we explain the main differences between the different profiles. To access C4I with your ESGF account and take advantange of all the C4I capabilities, please follow the instructions on our [Sign Up](https://dev.climate4impact.eu/c4i-frontend/signup) page.


<img src="images/home.png" alt="home" style="height:400px;width:800px;"/>



**NOTE:** C4I v2 is connected to the new ESGF authentication and authorisation infrastructure, whose development is currently being finalised by our partners. We apologise in advance for downtimes or instabilty of the login & authorisation functionality. Reliability of the service will improve soon. 

## How do I serach for data?

In the ***Data Discovery Page*** you select the project you want to access the data from (CMIP6 - CMIP5 - CORDEX).

Once you have selected the project, the portal will display a page where you can compose queries interactively by combining different properties. These are shown on the top of the page (*parameters*, *frequency*, *experiment*, *model*, *member*) and can be configured in any order.


<img src="images/search.png" alt="search" style="height:400px;width:800px;"/>

When configuring a specific property, to facilitate identifying the most relevant options, a subset of the possible values is be presented, grouped by different topics. This is the *Scoped* view. This presentation mode can be changed by switching the knob at the bottom left of the screen, from *Scoped* to *Extended*. The *Extended* presentation mode will display, for each property, all the possible values available for the current query configuration. 

In the image above we show that the summary of the current query is always displayed on the top of the screen. Each value can be interactively removed. Adding or removing a property value will always result in the live update of the  available choices for each property.

**NOTE:** at the time of writing this guide the authorisation to access CORDEX data is not in place yet. This is strictly dependant on the new authorisation infrastructure. We will enable this option as soon as possible.


## How do I select and download the data?

After you have prepared your query, C4I offers you more ways of accessing the data.

By pressing the ***Download*** button at the bottom of the ***Data Discovery Page*** page, C4I shows the list of datasets available in the ESGF that are matching your query. The list will be ordered by ***mirror***, so that you can easily find and select the data of interest from the same ESGF node. If needed, you can interactively change the default ordering.

<img src="images/select.png" alt="select" style="height:400px;width:800px;"/>

By clcking on the ***Link List*** button, a file containing the list of URLs associated with the data can be downloaded. The file is reprented in the standard [metalink](https://en.wikipedia.org/wiki/Metalink), which can be used to download the data independently in your processing scripts, adopting generic software libraries, or by using the interactive clients already available.

**NOTE:** Each dataset in the list is associated with a resolvable DOI page that describes the metadata of the model and the xperiment that produced the dataset. The page can be browsed by clicking on the file icon, under the DOI column. As an example, this [link](https://cera-www.dkrz.de/WDCC/ui/cerasearch/cmip6?input=CMIP6.ScenarioMIP.IPSL.IPSL-CM6A-LR.ssp585), corresponds to the page for the DOI *doi:10.22033/ESGF/CMIP6.5271* associated with all the data of the *IPSL-CM6A* model, *ssp585* experiment.

### How to prepare the data for analyses in C4I?

The ***Download*** page offers you the opportunity to access the new and experimental analysis functionalities of C4I. By clicking on ***Export To Notebook*** you request the portal to automatically export and manage the data onto an [*JupyterLab*](https://jupyterlab.readthedocs.io) instance, which is made available for you to use. 

<img src="images/getnotebook.png" alt="noteboook" style="height:400px;width:800px;"/>


C4I prepares the resources needed to provide you with a notebook enviroenment. This includes the execution of the workflows needed to copy the selected data (within the limit of the available storage) from the ESFG network to your storage, as well as the installation of the default *Python* modules that you will be using programmatically to develop your anaysis.(eg. *icclim, pandas, xarray, cartopy, etc.*).

This experimental environment is provided thanks to the SWIRRL API (Software for Reproducible Research Lab). The API delivers an enhanced vesion of *JupyterLab*, which is extended with a set of tools that allow you to **update**, **restore**, **share** and **reproduce** the progress of your research. You will find a guide on how to use SWIRRL![enter image description here](https://assets.gitlab-static.net/uploads/-/system/project/avatar/17988184/Screenshot_2020-04-09_at_10.37.38.png?width=32) directly in your notebook (Welcome to SWIRRL).


<img src="images/swirrl.png" alt="swirrl" style="height:400px;width:800px;"/>



**NOTE** Your *JupyterLab* instance may be accessible before all the data is copied to your storage. If you select more data that you can currently store, this will be notified and you can choose whether to create a new enivronment  with the new datasets (replacing the old one). 



<img src="images/resumenb.png" alt="resume" style="height:250px;width:600px;"/>


At the moment we do not support deletion or changes to the data already copied. This is to make sure your data storage is kept consistent and reproducible. However, you can choose to create a new enviroment at any time. For instance, if you want to start a new research, keeping the notebook focussed on your current interest. To allow the reproducibilty of past analysis in the future, we keep old data in a dedicate space, which is associated with snapshots of the progress of your developments (see the Welcome to SWIRRL guide - Snapshots!). 




### Are there examples to learn how to implement new analyses methods? 

To help with developing your own first analysis notebook, we have collected a number of examples. You can browse and download those you are interested in from the IS-ENES3 GitLab repository (https://gitlab.com/is-enes-cdi-c4i/notebooks). By selecting in C4I the data mentioned in the  sample notebooks, you can  export the files to your existing notebook or ask C4I to create a new one.  Upload the sample notebooks to your Jupyter Lab and, once the data has been copied (see the Welcome to SWIRRL guide for details), you will be able to test, change and experiment youself with this new experimental facility!





### Can I reduce the amount of data downloaded by subsetting?

Yes, this is possible. However not all of the ESFG nodes are supporting this type of service.
By clickin on the ***Subsetting*** button of the ***Data Disovery*** page, you can choose the datasets your are interetesed that match you query paramenters. By pressing the ***send to process*** button C4I will show you a configuration panel where you can set your subsetting parementes. This feature is currently linked to the experimental notebook facility, so the data will be made availabe for you in your Jupyter Lab instance. 

<img src="images/subsetting.png" alt="subsetting" style="height:350px;width:800px;"/>

In the IS-ENES project, we are working on implementing a more flexible and performing technology to enable subsetting and other types of typical data transformation operations. This will make the data smaller and more usable for you to proceed with futher analyses. We will integrate such operators in C4I in the coming releases!
